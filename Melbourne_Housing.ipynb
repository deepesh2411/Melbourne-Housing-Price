## price column

```python
q1 = mel_new["Price"].quantile(0.25)
q3 = mel_new["Price"].quantile(0.75)
iqr = q3-q1
floor = q1-(1.5*iqr)
cap = q3 + (1.5*iqr)
```

```python
mel_new[(mel_new["Price"]>= floor)&(mel_new["Price"]<= cap)]
```

```python
mel_new["Price"].count()
```

### Now doing for all columns accept rooms as we have done earlier

```python
mel_new.columns[0]        #fetching column name
```

```python
#for all column storing required value 

q1,q3,iqr,floor,cap = ([] for i in range(5))
for i in range(len(col_to_take)):                                
    q1_value = mel_new[mel_new.columns[i]].quantile(0.25)
    q3_value = mel_new[mel_new.columns[i]].quantile(0.75)
    iqr_value = q3_value-q1_value
    floor_value = q1_value - (1.5*iqr_value)
    cap_value = q3_value + (1.5*iqr_value)
    
    q1.append(q1_value)
    q3.append(q3_value)
    iqr.append(iqr_value)
    floor.append(floor_value)
    cap.append(cap_value)
    
print("Q1    :",q1)
print("Q3    :",q3)
print("iqr   :",iqr)
print("floor :",floor)
print("cap   :",cap)
```

```python
print (melbourne["Price"].count()," rows before outlier")
```

```python
#selecting rows btween the range for each column

for i in range(len(col_to_take)):
    mel_new = mel_new[(mel_new[mel_new.columns[i]] >= floor[i])&(mel_new[mel_new.columns[i]] <= cap[i])]
```

```python
print (mel_new["Price"].count()," rows after outlier")
```
